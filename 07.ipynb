## **Pandas**

```python
import pandas as pd
df = pd.read_csv('./student_data.csv')
```

```python
df.groupby('placement_status')['placement_status'].agg('count')
df.groupby('placement_status')['sleep_hours'].agg('mean')
df.groupby('placement_status').agg(mean=('placement_status','count'),avg=('sleep_hours','mean'))
```

```python
df.groupby('placement_status').agg(mean=('study_hours','mean'))
```

```python
df.groupby('study_hours').agg(placement=('placement_status','count'),Sleep=('sleep_hours','mean'))
```

```python
df.groupby('placement_status').agg(Mean=('exam_score','mean'))
```

```python
df.groupby('study_hours').agg(score=('exam_score','mean'))
```

```python
def grade(x):
  if x>90:
    return 'A'

  elif x>80:
    return 'B'

  elif x>70:
    return 'C'

  elif x>50:
    return 'D'

  elif x>35:
    return 'E'

  else :
    return 'F'

df['Grade'] = df['exam_score'].apply(grade)
df
```
