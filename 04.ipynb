# **NumPy**

```python
import numpy as np
```

```python
a = np.array([1,2,3,4,5,6, False])
b = np.array([1,2,3,4,5,6,"2"])
print(type(a))
print(a)
print(b)
```

```python
na = np.array([1,2,3,'hello',False])
print(bool(na[-1]))

a = False;
print(bool(a))
```

```python
a = np.zeros((2,3,4),dtype=int)     #by default it is float
print(a)

print(type(a[0][0][0]))
```

```python
a = np.ones((2,3,4),dtype=int)     #by default it is float
print(a)
```

```python
a = np.empty((2,3,4),dtype=int)     #by default it is float
print(a)
```

```python
a= np.full((2,3),5,dtype=int)
print(a)
```

```python
a= np.arange(2,50,2)
print(a)

al = np.linspace(2,10,5)
print(al)
```

```python
a = np.ones((3,4))
print(a.dtype)
print(a.size)
print(a.shape)
```

```python
a = np.random.randint(1,7,4) #start,stop,size
b = np.random.choice([1,2,3,4,5,6,7,8],2)

print(a)
print(b)
```

```python
a = np.array([1,4,6,5])
b = np.array([2,3,7,9])
# b = np.array([2,3,7,9,0])
print(a*2)
print(a+b)
print(a-b)
print(a/b)
print(a*b)

a=[1,2,3,4]
b=[5,6,7,8]
print(a*2)
```

```python
a = np.random.randint(2,15,6)
print(a)

b=a.reshape((2,3))
print(b)
```

```python
a = np.random.randint(2,15,9)
print(a)

b=a.reshape((3,3))
print(b)

c=b.transpose()
print(c)
```

```python
a = np.random.randint(2,15,9)
b = np.random.randint(2,15,8)

c = np.concatenate((a, b),axis=0)

print(c)
```

```python
a = np.random.randint(2,15,8)
b = np.random.randint(2,15,8)
print("A = " ,a)
print("B = ",b)

# c = np.where(condition , x , y) if true x otherwisw xy
c = np.where(b>a , b , a)
print(c)
```

```python
a = np.random.randint(2,15,8)
print(a)
print(np.sum(a))
# b= np.sort(a)
print(np.sort(a))
# c=np.mean(a)
print(np.mean(a))
# d=np.std(a)
print(np.std(a))

print(np.min(a))
print(np.max(a))

print(np.unique(a)) #sort with only unique elemnt
```

```python
a = np.random.randint(2,25,(4,4))
b = np.random.randint(2,25,(4,4))
print(a)
print(b)

print(np.matmul(a,b))
```
