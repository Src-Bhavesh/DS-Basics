```python
class Person:
    def __init__(self, h, w, age):
        self.height = h
        self.weight = w
        self.age = age

class Rahul(Person):
    def __init__(self, g, name, b, h, w, age):
        super().__init__(h, w, age)
        self.gender = g
        self.name = name
        self.bankbalance = b

a = Rahul("Male", "Rahul", 50000, 175, 70, 25)
```

```python
class complex :
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __add__(self,other):
    return complex(self.x+other.x,self.y+other.y)


  def display(self):
    print(self.x,"+i",self.y)


a = complex(2,3)
b = complex(4,5)

c=a+b

display()
```
