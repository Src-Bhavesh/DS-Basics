```python
import numpy as np
```

```python
a = np.random.randint(1,90,(5,5));
print(a)
print("Rowwise mean")
m1 = np.mean(a,axis=1);
print(m1)
print("Columnwise mean")
m2 = np.mean(a,axis=0);
print(m2)

w = np.where(a%2==0,-1,a)
print(w)

for i in range(len(a)):
  for j in range(len(a[i])):
    if(a[i][j]%2==0):
      a[i][j]=-1

print(a)

print("Rowwise Sort")
l1 = np.sort(a,axis=1);
print(l1)
print("Columnwise Sort")
l2 = np.sort(a,axis=0);
print(l2)
print("Sum Rowwise")
s1 = np.sum(a,axis=1);
print(s1)
print("sum rowwise")
s2 = np.sum(a,axis=0);
print(s2)
print("Count columnwise")
c1 = np.count_nonzero(a<0,axis=0)
print(c1)
c1 = np.sum(np.where(a<0,1,0),axis=0)
print(c1)




```

```python
stock = np.random.randint(30,100,(5,5))
print(stock)

demand = np.random.randint(1,100,(5,5))
print(demand)

remaining = stock-demand
print(remaining)

value = np.where(remaining<0,0,remaining)
print(value)

s=np.sum(remaining[remaining<0])
print("Total Shortage : ",s)

c = np.count_nonzero(value==0)
print("Product with zero stock : ",c)

st = np.sort(a,axis=1);
print(st)


```
