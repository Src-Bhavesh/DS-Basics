```python
import pandas as pd
```

```python

df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Raja','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]
})


df3 = pd.merge(df1,df2,on='id',how='inner')
df3.rename(columns={'name_x':'Name-1'},inplace=True)
print(df3)
print('\n')

df4 = pd.merge(df1,df2,on='id',how='left')
print(df4)
print('\n')

df5 = pd.merge(df1,df2,on='id',how='right')
print(df5)
print('\n')

df6 = pd.merge(df1,df2,on='id',how='outer')
print(df6 )
print('\n')

df7 = pd.merge(df1,df2,how='cross')
print(df7)
```

```python
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Raja','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]
})

df3 = pd.concat([df1,df2])   #should have same no. of column
print(df3)
print('\n')
```

```python
df = pd.DataFrame({
    'name':['a','d','c','a','b','d'],
    'age':[1,2,3,4,2,2]
})
print(df)

df=df.drop_duplicates(); #drop only is entire row is a duplicate
print(df)

df=df.drop_duplicates(subset=['name'],keep='first') #keep first or last
print(df)
```
