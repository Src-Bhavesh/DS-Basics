```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Line Chart**

```python
# x = np.linspace(0, 10, 100)
# y = np.sin(x)
# plt.plot(x,y)
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.figure(figsize=(5,5))  # Physical size of chart in inches
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Relation between number & Square")
plt.tight_layout()
plt.show()


```

## **Bar Graph**

```python
subjects = ["Maths","Physics","CS","English"]
scores = [88,76,95,80]

plt.bar(subjects,scores ,color='red',edgecolor = 'black')
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Bar Chart")
plt.show()

plt.bar(scores,subjects ,color='red',edgecolor = 'black')
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Bar Chart")
plt.show()
```

## **Scatter Graph**

```python
hours = np.random.randint(1,10,50);
marks = hours*8 + np.random.randint(1,20,50)
plt.scatter(hours,marks, color='red')
# plt.plot(hours,marks)
plt.show()
```

## **Histogram**

```python
data = np.random.normal(60, 10, 500)
plt.hist(data,color='red',bins=50,edgecolor="black") #Bins = no.of bars
plt.show()

data = np.random.randint(10,60, 500)
plt.hist(data,color='red',edgecolor="black")
plt.show()
```

## **PIE CHART**

```python
subjects = ["Maths","Physics","CS","English"]
scores = [30 ,20,40,10]

plt.pie(scores,labels=subjects,autopct="%.2f%%",shadow=True,startangle=90,explode=(0,0,0.2,0))
plt.tight_layout()
plt.show()
```

## **BoxPlot**

```python
batch1 = np.random.randint(10,70,100);
batch2 = np.random.randint(10,70,100);
plt.boxplot([batch1 , batch2],tick_labels=["batch1","batch2"] )

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("student_data.csv")
sns.lineplot(data=df,x="study_hours",y="exam_score")

```

```python
def grade(x):
  if x>90:
    return 'A'

  elif x>80:
    return 'B'

  elif x>70:
    return 'C'

  elif x>50:
    return 'D'

  elif x>35:
    return 'E'

  else :
    return 'F'

df['Grade'] = df['exam_score'].apply(grade)
sns.barplot(data=df,x="study_hours",y="Grade",estimator='median')
plt.title("Study Hours vs Grade")
plt.tight_layout()
plt.show()
```
