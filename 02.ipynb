### **List:**

```python
l = [1,2,3,'a','b',"Hello",[2,3]]
print(l)

# append()
l.append([1,2.3])
print(l)

#extend()
l.extend([1,2,3])
print(l)

# insert(idx,elem)
l.insert(2,"insert")
print(l)

# pop()
l.pop()
print(l)

# pop(idx)
l.pop(2)
print(l)

# remove(elem)
# l.remove("insert")
# print(l)

#index()
print(l.index(2))

#count()
print(l.count(2))

# sort
lst = [2,5,1,4,3,7,88,44,2,98,23]
lst.sort()
print(lst)
ls = sorted(lst)
print(ls)

# sum
print(sum(lst))

# min
print(min(lst))

# max
print(max(lst))

```
